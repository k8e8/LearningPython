```python
import matplotlib.pyplot as plt
labels = ['Asus', 'Samsung', 'Apple', 'Sony']
values = [20, 30, 45, 10]
colors =['gray', 'green', 'red', 'blue']
explode = (0, 0, 0.1, 0)

_, _, autotexts = plt.pie(values,labels=labels,colors=colors,explode=explode,shadow=True,startangle=-40,autopct='%1.1f%%')
plt.setp(autotexts, size=14, color='white')
plt.show()
```
